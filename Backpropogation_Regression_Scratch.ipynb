```python
# importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.DataFrame(data=[[8,8,4],[7,9,5],[6,10,6],[5,12,7]],columns=['cgpa','profile_score','lpa'])
```

```python
def initial_value(array):
    #[2,2,1]
    parameter={}
    for i in range(1,len(array)):
        parameter['W'+ str(i)]=np.ones((array[i-1],array[i]))*0.1
        parameter['B'+str(i)]=np.zeros((array[i],1))*0.1
    return parameter
```

```python
init_parameters=initial_value([2,2,1])
init_parameters
```

```python
X=data.iloc[:,:-1]
y=data.iloc[:,-1]
```

```python
# {'W1': array([[0.1, 0.1],
#         [0.1, 0.1]]),
#  'B1': array([[0.],
#         [0.]]),
#  'W2': array([[0.1],
#         [0.1]]),
#  'B2': array([[0.]])}
    
A1={}
A2={}
def calculate_prediction(parameters,X):
#     for i in range(len(init_parameters)):
    i=0
    w1=parameters['W1'][i][i]
    w2=parameters['W1'][i][i+1]
    w3=parameters['W1'][i+1][i]
    w4=parameters['W1'][i+1][i+1]
    w5=parameters['W2'][i][i]
    w6=parameters['W2'][i+1][i]
    b1=parameters['B1'][i][i]
    b2=parameters['B1'][i+1][i]
    b3=parameters['B2'][i][i]
#         A1={'W1':w1,'W2':w2,'cgpa':cgpa,'profile_score':profile_score,'B1':b1,'B2':b2}
    A1['o1']=w1*X[0]+w2*X[1]+b1
    A1['o2']=w3*X[0]+w4*X[1]+b2
    A2['o3']=np.round(w5*A1['o1']+w6*A1['o2']+b3,2)
    return A1,A2
```

```python
# {'W1': array([[0.1, 0.1],
#         [0.1, 0.1]]),
#  'B1': array([[0.],
#         [0.]]),
#  'W2': array([[0.1],
#         [0.1]]),
#  'B2': array([[0.]])}
def updating_parameters(parameters,y,y_hat,x,A1):
    i=0
    parameters['W1'][i][i]=parameters['W1'][i][i] + (0.001*2*(y-y_hat)*parameters['W2'][i][i]*x[i])
    parameters['W1'][i][i+1]=parameters['W1'][i][i+1] + (0.001*2*(y-y_hat)*parameters['W2'][i][i]*x[i+1])
    parameters['B1'][i][i]=parameters['B1'][i][i] + (0.001*2*(y-y_hat)*parameters['W2'][i][i])
    
    parameters['W1'][i+1][i]=parameters['W1'][i+1][i] + (0.001*2*(y-y_hat)*parameters['W2'][i+1][i]*x[i])
    parameters['W1'][i+1][i+1]=parameters['W1'][i+1][i+1] + (0.001*2*(y-y_hat)*parameters['W2'][i+1][i]*x[i+1])
    parameters['B1'][i+1][i]=parameters['B1'][i+1][i] + (0.001*2*(y-y_hat)*parameters['W2'][i+1][i])
    
    parameters['W2'][i][i]=parameters['W2'][i][i] + (0.001*2*(y-y_hat)*A1['o1'])
    parameters['W2'][i+1][i]=parameters['W2'][i+1][i] + (0.001*2*(y-y_hat)*A1['o2'])
    parameters['B2'][i][i]=parameters['B2'][i][i] + (0.001*2*(y-y_hat))
    
    return parameters
```

```python
# implementing epochs

init_parameters=initial_value([2,2,1]) # providing initial values to the weights and baises
x=[]
Loss = []
updated_parameters={}
for i in range(100):
    for j in range(data.shape[0]):
        x.append(X.iloc[j][0])
        x.append(X.iloc[j][1])
#         print(init_parameters)
        A1,y_hat=calculate_prediction(init_parameters,x)
        updated_parameters=updating_parameters(init_parameters,y[j],y_hat['o3'],x,A1)
        init_parameters=updated_parameters
#         print(A1)
        print(y_hat['o3'])
        Loss.append((y-y_hat['o3'])**2)

    print('Epoch - ',i+1,'Loss - ',np.array(Loss).mean())
```
